```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def PlotData(fileName):
    electricity_household = pd.read_csv(fileName + '.tsv',\
        sep='\t',header=0,)
    electricity_household = electricity_household.set_index('country')
    # print(electricity_household.dtypes)
    # print(electricity_household.iloc[2][65])

    fig = plt.figure(dpi=500)
    electricity_household.T.plot(ax=plt.gca())
    plt.title(fileName)
    plt.savefig(fileName + '.png')
```

```python
PlotData('electricity_household')
PlotData('electricity_industry')
PlotData('gas_household')
PlotData('gas_industry')
```
